```python
import numpy as np
```

```python
df_num = df_houseprice.select_dtypes(include = [np.number])
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_target = df_houseprice['Price']

df_num =df_num.drop('Price' , axis =1)
```

```python
#iniciate  the standard scaler
ss = StandardScaler()
```

```python
#tranform the data
scale = ss.fit_transform(df_num)
```

```python
scale
```

```python
scale = pd.DataFrame(scale , columns= df_num.columns)
scale = scale.reset_index()
```

```python
scale.head()
```

```python
#make dataframe called independent variable
df_independent = pd.concat([scale , dummy_var] , axis = 1)

df_independent.head()
```

```python
# split  for the  training and training

```

```python
import statsmodels
import statsmodels.api as sm

df_independent = sm.add_constant(df_independent)
```

```python
df_independent
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train , y_test = train_test_split(df_independent ,df_target, random_state=10, test_size = 0.2)                                                                                                 
```

```python
print('x_train' , x_train.shape)
print('y_train' , y_train.shape)

print('x_test' , x_test.shape)
print('x_test' , y_test.shape)
```

# linear Regression


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#create linear regression
model = LinearRegression()

#fit the model to the training data 
model.fit(x_train , y_train)

#predict on the testing data
prediction = model.predict(x_test)
```

```python
prediction
```

```python
mse = mean_squared_error(y_test , prediction)
print("mean_squared_error:" , mse)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test , prediction)
print("mean_absolute_error :", mae)
```
